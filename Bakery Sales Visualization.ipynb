# Deleting Rows Syntax :



```python
import pandas as pd
```

```python
df = pd.read_csv("bakerylinesales.csv")
```

```python
df
```

```python
# Data cleaning


df.drop(labels=[7, 8], axis = 0)
```

```python
df['dates'] = pd.to_datetime(df['dates'])
```

```python
df['dates'].dtypes
```

```python
new_df = df.drop(columns=['Unnamed: 1'], axis = 1)
```

```python
my_new_df = new_df.drop(labels=[7, 8], axis = 0)
```

```python
my_new_df
```

```python
#removing approx

my_new_df['total'] = my_new_df['total'].str.strip('approx')

#changing dtype

my_new_df['total'] = my_new_df['total'].astype(int)
```

```python
my_new_df['total']
```

```python
import matplotlib.pyplot  as plt
```

```python
# Setting X , Y Axis
plt.plot(my_new_df["dates"] , my_new_df['total'])

# Heading
plt.title('Monthly Total Sales', size = 16)
# X Label
plt.xlabel('Date')
# Y Label
plt.ylabel('Total Sales (Mon)')

# Removing Hex Code
plt.show()
```

```python
# Theme
plt.style.use('grayscale')

# Figure Size
plt.figure(figsize = (10,4))

# Setting X , Y Axis
plt.plot(my_new_df["dates"] , my_new_df['total'] , color = 'skyblue', linewidth = 2)

# Heading
plt.title('Monthly Total Sales', size = 16)
# X Label
plt.xlabel('Date')
# Y Label
plt.ylabel('Total Sales (Mon)')

# for removing exponent values in y axis
plt.ticklabel_format(axis = "y", style = "plain")

# Removing Hex Code
plt.show()

```
